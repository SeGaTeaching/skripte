#### **4.1 Das Prozedurale Paradigma (Unser bisheriger Weg)**

**Definition**
Das prozedurale Paradigma ist ein Programmierstil, der ein Programm als eine logische Abfolge von Anweisungen, Prozeduren oder Funktionsaufrufen strukturiert. Der Kern dieses Paradigmas ist die strikte Trennung von **Daten**, auf denen operiert wird, und den **Prozeduren (Funktionen)**, die diese Operationen ausf√ºhren.

Man kann sich das so vorstellen: Die Daten sind passive, leblose "Zutaten" (wie Mehl, Zucker, Eier), die in verschiedenen Variablen gespeichert werden. Die Funktionen sind die aktiven "Rezepte" aus einem separaten Kochbuch, die diese Zutaten entgegennehmen, verarbeiten und ein Ergebnis produzieren.

**Probleme und Grenzen des prozeduralen Ansatzes**
F√ºr kleine und mittlere Projekte ist dieser Ansatz einfach und effektiv. Bei gro√üen, komplexen Anwendungen st√∂√üt er jedoch an seine Grenzen, was zu schwer wartbarem Code f√ºhrt. Dieses Ph√§nomen wird oft als **"Softwarekrise"** der 1970er und 80er Jahre bezeichnet, die zur Entwicklung von OOP f√ºhrte.

Die Hauptprobleme sind:
1.  **Globaler Zustand (Global State):** In gro√üen prozeduralen Programmen gibt es oft viele globale Variablen, auf die eine Vielzahl von Funktionen zugreifen und die sie ver√§ndern k√∂nnen. Wenn ein Fehler auftritt, ist es extrem schwierig nachzuvollziehen, welche Funktion zu welchem Zeitpunkt die Daten unerw√ºnscht ver√§ndert hat. Man verliert den √úberblick, wer f√ºr welchen Datenzustand verantwortlich ist. Dies f√ºhrt zu unvorhersehbaren **Seiteneffekten (Side Effects)**.

2.  **Geringe Wiederverwendbarkeit & starke Kopplung:** Eine Funktion wie `berechneGehalt(mitarbeiterDaten)` ist eng an eine ganz bestimmte Datenstruktur (`mitarbeiterDaten`) gekoppelt. √Ñndert sich die Struktur der Daten (z.B. ein neues Feld kommt hinzu), m√ºssen potenziell **alle** Funktionen, die diese Daten verwenden, angepasst werden. Es ist schwierig, eine Funktion aus einem Projekt herauszul√∂sen und in einem anderen wiederzuverwenden, weil die dazugeh√∂rigen Datenstrukturen fehlen.

3.  **Schwierige Abbildung der Realit√§t:** Die reale Welt besteht nicht aus getrennten Daten und Funktionen. Ein **Auto** *hat* eine Farbe und eine PS-Zahl (Daten) und es *kann* fahren und hupen (Funktionen). Das prozedurale Modell zwingt uns, diese Einheit k√ºnstlich zu trennen: hier die Variable `autoDaten`, dort die globale Funktion `fahre(autoDaten)`. Das ist nicht intuitiv und erschwert die Modellierung komplexer Systeme.

#### **4.2 Das Objektorientierte Paradigma (Ein neuer Denkansatz)**

Die **Objektorientierte Programmierung (OOP)** ist keine neue Technik, sondern ein fundamental anderer **Denkansatz (Paradigma)**. Die Kernidee von OOP ist die √úberwindung der Trennung von Daten und Funktionen durch das Konzept der **Kapselung**.

**Die Kernidee: Objekte**
In der OOP b√ºndeln wir zusammengeh√∂rige Daten und die Funktionen, die exklusiv auf diesen Daten arbeiten, in einer einzigen, logischen und abgeschlossenen Einheit. Diese Einheit nennen wir ein **Objekt**.

* Die Daten eines Objekts nennt man **Attribute** (oder Eigenschaften, Member-Variablen).
* Die Funktionen eines Objekts nennt man **Methoden**.

Ein Objekt ist also ein "intelligenter" Datencontainer. Es verwaltet seinen eigenen Zustand (seine Attribute) und bietet nach au√üen hin √ºber seine Methoden kontrollierte "Dienstleistungen" an. Anstatt einer globalen `fahre(auto)`-Funktion gibt es jetzt ein `auto`-Objekt mit einer `fahren()`-Methode, die man aufruft: `meinAuto.fahren()`.

**Die drei S√§ulen der OOP**
Das gesamte OOP-Paradigma ruht auf drei fundamentalen S√§ulen, die wir im n√§chsten Block im Detail lernen werden, hier aber schon konzeptionell verstehen m√ºssen:

1.  **Kapselung (Encapsulation):** Dies ist das bereits erw√§hnte B√ºndeln von Attributen und Methoden in einem Objekt. Ein wichtiger Teil davon ist das **Information Hiding**: Das Objekt sch√ºtzt seine internen Daten vor unkontrolliertem Zugriff von au√üen. Man kann den Zustand eines Objekts nur √ºber seine √∂ffentlichen Methoden √§ndern. Man kann einem Auto-Objekt nicht einfach sagen "deine Geschwindigkeit ist jetzt 500", man muss die Methode `beschleunige()` aufrufen, die vielleicht interne Limits pr√ºft.

2.  **Vererbung (Inheritance):** Erm√∂glicht es, neue **Klassen** (Baupl√§ne f√ºr Objekte) auf Basis von bestehenden Klassen zu erstellen. Eine Klasse `LKW` kann von einer allgemeinen Klasse `Fahrzeug` *erben*. Der `LKW` erh√§lt dadurch automatisch alle Attribute (`ps`, `farbe`) und Methoden (`fahren()`, `bremsen()`) des `Fahrzeugs` und kann zus√§tzliche, eigene hinzuf√ºgen (z.B. `ladeflaeche_beladen()`). Dies f√∂rdert die Wiederverwendbarkeit von Code massiv.

3.  **Polymorphismus (Vielgestaltigkeit):** Bedeutet, dass unterschiedliche Objekte auf dieselbe Nachricht (denselben Methodenaufruf) auf ihre eigene, spezifische Weise reagieren k√∂nnen. Ein Aufruf `zeichne()` k√∂nnte bei einem `Kreis`-Objekt einen Kreis und bei einem `Rechteck`-Objekt ein Rechteck auf den Bildschirm malen. Das Programm muss nicht wissen, um welches Objekt es sich genau handelt, es sendet nur die Nachricht "zeichne dich!".

## **Beispiel 1: Der Hund**

Ein sehr intuitives Beispiel, um die B√ºndelung von Daten und Verhalten zu zeigen.

#### **Prozeduraler Ansatz**

**Die Logik:**
Die Daten des Hundes (sein Name und seine Energie) werden in separaten Variablen gespeichert. Die Aktionen (Bellen, Fressen) sind allgemeine, globale Funktionen, denen man die Daten des Hundes √ºbergeben muss, damit sie damit arbeiten k√∂nnen.

**Code:**

```python
# --- DATEN ---
hund_name = "Bello"
hund_energie = 5

# --- FUNKTIONEN ---
def bellen(name_des_hundes):
    print(f"{name_des_hundes} bellt: Wuff!")

def fressen(aktuelle_energie):
    print("Der Hund frisst...")
    aktuelle_energie += 2
    return aktuelle_energie

def spielen(aktuelle_energie):
    print("Der Hund spielt...")
    aktuelle_energie -= 3
    return aktuelle_energie

# --- PROGRAMMABLAUF ---
bellen(hund_name)
hund_energie = spielen(hund_energie)
print(f"Bello hat jetzt {hund_energie} Energie.")
hund_energie = fressen(hund_energie)
print(f"Bello hat jetzt {hund_energie} Energie.")
```

**Erkl√§rung f√ºr den Unterricht:**

  * Zeige, wie **Daten und Funktionen komplett getrennt** sind. Wir haben hier die Variablen und dort die Funktionen.
  * Betone, dass die Funktionen **nicht von sich aus wissen**, auf welchen Hund sie sich beziehen. Wir m√ºssen die Daten (`hund_name`, `hund_energie`) bei jedem Aufruf explizit **√ºbergeben**.
  * Weise darauf hin, dass wir den Energiewert nach dem Aufruf von `fressen` oder `spielen` **manuell aktualisieren** m√ºssen (`hund_energie = fressen(...)`).

#### **Objektorientierter Ansatz (OOP)**

**Die Logik:**
Wir erstellen einen "Bauplan" (eine **Klasse**) namens `Hund`. Jeder konkrete Hund, den wir erstellen (ein **Objekt**), kapselt seine eigenen Daten (Attribute wie `name`, `energie`) und seine eigenen F√§higkeiten (Methoden wie `bellen`, `fressen`) in sich.

**Code:**

```python
# --- BAUPLAN (KLASSE) ---
class Hund:
    def __init__(self, name, energie=5):
        # Attribute (die Daten des Objekts)
        self.name = name
        self.energie = energie

    # Methoden (die F√§higkeiten des Objekts)
    def bellen(self):
        print(f"{self.name} bellt: Wuff!")

    def fressen(self):
        print(f"{self.name} frisst...")
        self.energie += 2

    def spielen(self):
        print(f"{self.name} spielt...")
        self.energie -= 3

# --- PROGRAMMABLAUF ---
bello = Hund("Bello") # Erstellen eines Objekts (Instanz)

bello.bellen()
bello.spielen()
print(f"{bello.name} hat jetzt {bello.energie} Energie.")
bello.fressen()
print(f"{bello.name} hat jetzt {bello.energie} Energie.")
```

**Erkl√§rung f√ºr den Unterricht:**

  * Hier sind **Daten und Funktionen in einer Einheit (dem Objekt `bello`) geb√ºndelt**.
  * Die Methode `bello.fressen()` **wei√ü automatisch**, auf welchen Energielevel sie sich bezieht ‚Äì n√§mlich auf den von `bello`. Wir m√ºssen nichts √ºbergeben.
  * Der Zustand (`bello.energie`) wird **innerhalb des Objekts selbst** verwaltet. Der Aufruf `bello.fressen()` ver√§ndert das Objekt direkt, wir brauchen kein manuelles Update von au√üen.
  * Das ist **intuitiver und n√§her an der Realit√§t**: Ein Hund "hat" einen Namen und "kann" bellen.

### **Beispiel 2: Das Bankkonto**

Ein perfektes Beispiel, um das Prinzip der Kapselung und Datensicherheit zu zeigen.

#### **Prozeduraler Ansatz**

**Die Logik:**
Der Kontostand ist eine einfache globale Variable. Jede Funktion kann direkt darauf zugreifen und sie ver√§ndern.

**Code:**

```python
# --- DATEN (Globaler Zustand) ---
kontostand = 1000.00

# --- FUNKTIONEN ---
def einzahlen(betrag):
    global kontostand # Zugriff auf die globale Variable n√∂tig
    kontostand += betrag
    print(f"Einzahlung erfolgreich. Neuer Kontostand: {kontostand}‚Ç¨")

def abheben(betrag):
    global kontostand
    if betrag <= kontostand:
        kontostand -= betrag
        print(f"Abhebung erfolgreich. Neuer Kontostand: {kontostand}‚Ç¨")
    else:
        print("Fehler: Kontodeckung nicht ausreichend.")

# --- PROGRAMMABLAUF ---
einzahlen(200)
abheben(500)
# Problem: Direkter, unkontrollierter Zugriff ist m√∂glich!
kontostand = -5000 # Unrealistischer Zustand, aber technisch m√∂glich
print(f"Manipulierter Kontostand: {kontostand}‚Ç¨")
```

**Erkl√§rung f√ºr den Unterricht:**

  * Das Hauptproblem hier ist der **unkontrollierte Zugriff**. Jede Funktion (und jeder andere Teil des Programms) kann die globale Variable `kontostand` direkt manipulieren.
  * Wir k√∂nnten versehentlich einen negativen Kontostand setzen, ohne dass eine Pr√ºflogik greift. Die Daten sind **ungesch√ºtzt**.

#### **Objektorientierter Ansatz (OOP)**

**Die Logik:**
Die Klasse `Konto` kapselt den Kontostand. Er kann von au√üen nicht direkt ver√§ndert werden. Jede √Ñnderung muss √ºber die kontrollierten Methoden `einzahlen` und `abheben` laufen, die eine Pr√ºflogik enthalten.

**Code:**

```python
# --- BAUPLAN (KLASSE) ---
class Konto:
    def __init__(self, inhaber, startguthaben=0):
        self.inhaber = inhaber
        # Das Attribut ist "privat" und gesch√ºtzt
        self.__kontostand = startguthaben

    def einzahlen(self, betrag):
        if betrag > 0:
            self.__kontostand += betrag
            self.kontostand_anzeigen()

    def abheben(self, betrag):
        if betrag > 0 and betrag <= self.__kontostand:
            self.__kontostand -= betrag
            self.kontostand_anzeigen()
        else:
            print("Fehler: Abhebung nicht m√∂glich.")

    def kontostand_anzeigen(self):
        print(f"Neuer Kontostand f√ºr {self.inhaber}: {self.__kontostand}‚Ç¨")

# --- PROGRAMMABLAUF ---
mein_konto = Konto("Max Mustermann", 1000)

mein_konto.einzahlen(200)
mein_konto.abheben(500)
# Problem: Direkter Zugriff ist nicht mehr (so einfach) m√∂glich!
# mein_konto.__kontostand = -5000 # F√ºhrt nicht zum gew√ºnschten Ergebnis
# print(f"Kontostand bleibt gesch√ºtzt.")
mein_konto.kontostand_anzeigen()
```

**Erkl√§rung f√ºr den Unterricht:**

  * Das ist **Kapselung** in Reinform. Das Objekt ist der W√§chter seiner eigenen Daten.
  * Wir k√∂nnen den Kontostand nicht mehr direkt manipulieren. Wir m√ºssen eine "Anfrage" an das Objekt stellen, indem wir eine seiner Methoden aufrufen (z.B. `mein_konto.abheben(500)`).
  * Die Methode selbst enth√§lt die **Logik zur Validierung**. Sie entscheidet, ob die Zustands√§nderung erlaubt ist. Das macht den Code **sicher und robust**.

### **Beispiel 3: Das Rechteck**

Ein einfaches geometrisches Beispiel, das zeigt, wie Objekte ihren eigenen Zustand kennen.

#### **Prozeduraler Ansatz**

**Die Logik:**
Die Dimensionen eines Rechtecks werden in losen Variablen gehalten. Funktionen zur Berechnung von Fl√§che oder Umfang ben√∂tigen diese Dimensionen bei jedem Aufruf als Parameter.

**Code:**

```python
# --- DATEN ---
rechteck_breite = 10
rechteck_hoehe = 5

# --- FUNKTIONEN ---
def berechne_flaeche(b, h):
    return b * h

def berechne_umfang(b, h):
    return 2 * (b + h)

# --- PROGRAMMABLAUF ---
flaeche = berechne_flaeche(rechteck_breite, rechteck_hoehe)
umfang = berechne_umfang(rechteck_breite, rechteck_hoehe)

print(f"Die Fl√§che betr√§gt: {flaeche}")
print(f"Der Umfang betr√§gt: {umfang}")
```

**Erkl√§rung f√ºr den Unterricht:**

  * Die Funktion `berechne_flaeche` hat **keine Ahnung**, zu welchem Rechteck sie geh√∂rt. Sie ist eine allgemeine mathematische Funktion.
  * Wir m√ºssen die Daten (`rechteck_breite`, `rechteck_hoehe`) und die Funktion (`berechne_flaeche`) im Hauptprogramm manuell **zusammenf√ºhren**.

#### **Objektorientierter Ansatz (OOP)**

**Die Logik:**
Ein `Rechteck`-Objekt kennt seine eigene Breite und H√∂he. Die Methoden zur Berechnung von Fl√§che und Umfang sind Teil des Objekts und greifen auf dessen eigene, interne Daten zu.

**Code:**

```python
# --- BAUPLAN (KLASSE) ---
class Rechteck:
    def __init__(self, breite, hoehe):
        self.breite = breite
        self.hoehe = hoehe

    def berechne_flaeche(self):
        # Greift auf die eigenen Attribute zu
        return self.breite * self.hoehe

    def berechne_umfang(self):
        return 2 * (self.breite + self.hoehe)

# --- PROGRAMMABLAUF ---
mein_rechteck = Rechteck(10, 5)

# Wir fragen das Objekt nach seinen Eigenschaften
flaeche = mein_rechteck.berechne_flaeche()
umfang = mein_rechteck.berechne_umfang()

print(f"Die Fl√§che betr√§gt: {flaeche}")
print(f"Der Umfang betr√§gt: {umfang}")
```

**Erkl√§rung f√ºr den Unterricht:**

  * Der Code ist **intuitiver und lesbarer**. Wir rufen nicht eine allgemeine Funktion auf, sondern wir fragen das konkrete `mein_rechteck`-Objekt: "Berechne *deine* Fl√§che\!".
  * Die Methode `berechne_flaeche` ben√∂tigt **keine Parameter**, weil das Objekt, zu dem es geh√∂rt, seine eigene Breite und H√∂he bereits kennt.
  * Wenn wir ein zweites Rechteck (`anderes_rechteck = Rechteck(20, 30)`) erstellen, hat dieses seinen eigenen, unabh√§ngigen Zustand. `anderes_rechteck.berechne_flaeche()` w√ºrde ein v√∂llig anderes Ergebnis liefern.
