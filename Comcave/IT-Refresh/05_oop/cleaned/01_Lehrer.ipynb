#### **1.1 Klassen & Objekte (Der Bauplan & das Produkt)**

In der prozeduralen Programmierung haben wir Daten und Funktionen getrennt. Die Objektorientierte Programmierung (OOP) f√ºhrt diese beiden Welten zusammen. Die grundlegenden Bausteine daf√ºr sind **Klassen** und **Objekte**.

Stellt euch eine **Klasse** wie einen detaillierten **Bauplan** oder eine Pl√§tzchenform vor. Der Plan legt alle Eigenschaften und F√§higkeiten fest, die ein sp√§teres Produkt haben wird, aber er ist noch nicht das Produkt selbst.

Ein **Objekt** ist das tats√§chliche, konkrete **Produkt**, das nach diesem Bauplan erstellt wird. Es ist die Instanz einer Klasse. Aus einer einzigen Pl√§tzchenform (Klasse) k√∂nnen wir viele individuelle, einzigartige Pl√§tzchen (Objekte) ausstechen. Jedes Objekt existiert unabh√§ngig von den anderen, auch wenn sie nach demselben Plan gebaut wurden.

#### **1.2 Attribute & Konstruktor (Die Eigenschaften & die "Geburt")**

Ein Bauplan ist nutzlos, wenn er keine Eigenschaften beschreibt. In OOP nennt man die Eigenschaften eines Objekts **Attribute** (auch Instanzvariablen). Sie sind das "Ged√§chtnis" oder der **Zustand** eines Objekts.

Damit ein Objekt bei seiner Erschaffung direkt mit sinnvollen Anfangswerten ausgestattet wird, gibt es eine spezielle Methode: den **Konstruktor**. Er ist der "Geburtshelfer" f√ºr ein Objekt und wird genau einmal aufgerufen, wenn ein neues Objekt mit dem `new`-Schl√ºsselwort instanziiert wird.

Innerhalb des Konstruktors verwenden wir das Schl√ºsselwort `this`. `this` ist ein Verweis auf **"dieses konkrete Objekt, das gerade erschaffen wird"**. Damit unterscheiden wir zwischen dem Parameter, der von au√üen kommt (z.B. `farbe`), und dem internen Attribut des Objekts (`this.farbe`).

**Inkrementelles Projekt "Auto", Schritt 2: Eigenschaften hinzuf√ºgen**
Wir erweitern unseren Bauplan um Attribute und einen Konstruktor, der diese bei der "Geburt" eines Autos festlegt.

```java
// Java
public class Auto {
    // 1. Attribute definieren
    String marke;
    String farbe;
    int ps;

    // 2. Konstruktor definieren
    public Auto(String marke, String farbe, int ps) {
        // 'this.marke' ist das Attribut des Objekts.
        // 'marke' ist der Parameter, der von au√üen kommt.
        this.marke = marke;
        this.farbe = farbe;
        this.ps = ps;
    }
}

// In der main-Methode: Objekte mit spezifischen Eigenschaften erstellen
Auto meinVW = new Auto("Volkswagen", "Blau", 150);
Auto deinBMW = new Auto("BMW", "Schwarz", 220);

System.out.println(meinVW.marke); // Gibt "Volkswagen" aus
System.out.println(deinBMW.marke); // Gibt "BMW" aus
```

#### **1.3 Methoden (Die F√§higkeiten)**

Ein Objekt hat nicht nur Eigenschaften (Attribute), sondern auch F√§higkeiten (Verhalten). In OOP nennen wir diese F√§higkeiten **Methoden**. Methoden sind Funktionen, die zu einer Klasse geh√∂ren und auf die Attribute des eigenen Objekts zugreifen und diese ver√§ndern k√∂nnen.

Anstatt eine globale Funktion `druckeInfo(meinVW)` aufzurufen, bitten wir das Objekt, seine F√§higkeit selbst auszuf√ºhren: `meinVW.getInfoString()`. Die Methode wei√ü dank `this` automatisch, auf welche Daten (die von `meinVW`) sie zugreifen soll.

**Inkrementelles Projekt "Auto", Schritt 3: F√§higkeiten hinzuf√ºgen**
Unser Auto lernt jetzt zu hupen und seine Informationen preiszugeben.

```java
// Java
public class Auto {
    String marke;
    String farbe;
    int ps;

    public Auto(String marke, String farbe, int ps) {
        this.marke = marke;
        this.farbe = farbe;
        this.ps = ps;
    }

    // Neue Methode, die die Attribute des Objekts nutzt
    public String getInfoString() {
        return "Marke: " + this.marke + ", Farbe: " + this.farbe + ", PS: " + this.ps;
    }

    // Neue Methode, die eine einfache Aktion ausf√ºhrt
    public void hupen() {
        System.out.println("Hup! Hup!");
    }
}

// In der main-Methode: Methoden aufrufen
Auto meinVW = new Auto("Volkswagen", "Blau", 150);
meinVW.hupen(); // Gibt "Hup! Hup!" aus
System.out.println(meinVW.getInfoString()); // Gibt "Marke: Volkswagen, ..." aus
```
